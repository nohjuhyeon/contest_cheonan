# 신혼부부의 소득 구간별 출산자녀 현황

## 분석 목적 : 소득구간별 신혼부부의 출산율 확인

## 분석 방향
1. 소득구간별 신혼부부의 출산율 비교
2. 소득구간별 신혼부부의 출산율을 시간에 따른 변화 확인



## 주요 용어 정리
- 신혼부부 : 	통계작성 기준일(매년 11월 1일) 현재 혼인신고(가족관계의 등록 등에 관한 법률에 의한) 후 5년이 경과하지 않은 부부
- 소득 : 일을 통해 벌어들이는 연간 소득(근로.사업소득)에 대해 부부 합산액을 작성

### data 불러오기

```python
# 출산율 data 불러오기

# CSV 파일 경로 지정
file_path01 = '../..//dataset/Newborns_by_Newlywed_Income_Bracket/Newborns_by_Newlywed_Income_Bracket.csv'

# CSV 파일을 DataFrame으로 읽어오기
df_birthrate_income_range = pd.read_csv(file_path01, encoding='euc-kr')
df_birthrate_income_range
```

# 데이터 전처리

```python
df_birthrate_income_range['소득구간별(1)'] = df_birthrate_income_range['소득구간별(1)'].str.replace('천','000')
df_birthrate_income_range['소득구간별(1)'] = df_birthrate_income_range['소득구간별(1)'].str.replace('억','0000만')
```

```python
df_birthrate_income_range.rename(columns={'데이터':'차지 비율'},inplace=True)
```

# 2022년도 소득구간별 자녀가 있는 신혼부부의 차지 비율 확인

```python
df = df_birthrate_income_range[(df_birthrate_income_range['출산자녀수별(2)']=='소계')&(df_birthrate_income_range['출산자녀수별(1)']=='자녀 있음')&(df_birthrate_income_range['시점']==2022)]
df_pivot = pd.pivot_table(df,index='소득구간별(1)',values=['차지 비율'])
df_pivot = df_pivot.loc[['1000만원 미만','1000만원~3000만원 미만','3000만원~5000만원 미만','5000만원~7000만원 미만','7000만원~10000만원 미만','10000만원 이상'],:]
plt.figure(figsize=(10, 6))
plt.bar(df_pivot.index, df_pivot['차지 비율'], label='차지 비율')

# 추세선 추가
x = np.arange(len(df_pivot.index))
y = df_pivot['차지 비율'].values

# 레이블 및 제목 설정
plt.xlabel('소득 구간')
plt.ylabel('차지 비율')
plt.title('소득 구간별 차지 비율')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# 연도별 자녀가 있는 신혼부부의 차지 비율 확인

```python
df = df_birthrate_income_range[(df_birthrate_income_range['출산자녀수별(2)']=='소계')&(df_birthrate_income_range['출산자녀수별(1)']=='자녀 있음')]
df_pivot = pd.pivot_table(df,index='시점',columns = ['소득구간별(1)'],values=['차지 비율'])
df_pivot.plot(kind='line',figsize=(12,8))
plt.xlabel('시점')
plt.ylabel('자녀가 있는 신혼부부의 차지 비율')
plt.title('연도별 자녀가 있는 신혼부부의 차지 비율')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```

# 결론
- 전체적으로 상대적으로 소득이 적은 신혼부부가 자녀가 있을 확률이 더 높음
- 자녀가 있는 신혼부부의 연령대를 파악하여 소득이 적은 신혼부부에서 더 자녀가 있을 확률이 높은 이유에 대한 추가적인 분석이 필요
