# 연령별 출생아 부모 중 육아 휴직 대상자와 육아 휴직자 비교

## 분석 목적 : 연령별 출생아 부모 중 육아 휴직 대상자 중 실제 육아 휴직자 확인

## 분석 방향
1. 연령별 출생아 부모 중 육아 휴직 대상자 확인
2. 연령별 출생아 부모 중 육아 휴직자 확인
3. 휴직 대상자 중 실제 육아 휴직자 차지 비율 확인



## 주요 용어 정리
- 육아휴직자 수 : 만 8세 또는 초등학교 2학년 이하의 자녀를 대상으로 육아휴직을 시작한 부모의 수
- 육아휴직자 사용률 : (t년도 출생아 부모 중 t년도에 육아휴직을 사용한 사람/t년도 출생아 부모 중 육아휴직 대상조건을 충족시키는 사람) * 100


### data 불러오기

```python
# 출산율 data 불러오기

# CSV 파일 경로 지정
file_path01 = '../../dataset/Parental_Leave_Usage_Rate_by_Parents_Age/Parental_Leave_Usage_by_Parents_Age.csv'
file_path02 = '../../dataset/Parental_Leave_Usage_Rate_by_Parents_Age/Parental_Leave_Eligible_by_Parents_Age.csv'
file_path03 = '../../dataset/Parental_Leave_Usage_Rate_by_Parents_Age/Parental_Leave_Usage_Rate_by_Parents_Age.csv'
# CSV 파일을 DataFrame으로 읽어오기
parental_leave = pd.read_csv(file_path01, encoding='euc-kr')
parental_leave_eligible = pd.read_csv(file_path02, encoding='euc-kr')
parental_leave_ratio = pd.read_csv(file_path03, encoding='euc-kr')
```

# 데이터 전처리

```python
# 2022년 데이터는 잠정치이기 때문에 제외
parental_leave = parental_leave.loc[parental_leave['시점']!='2022 p)']
parental_leave_eligible = parental_leave_eligible.loc[parental_leave_eligible['시점']!='2022 p)']
parental_leave_ratio = parental_leave_ratio.loc[parental_leave_ratio['시점']!='2022 p)']
```

```python
parental_leave.rename(columns={'데이터':'육아 휴직자'},inplace=True)
parental_leave_eligible.rename(columns={'데이터':'육아 휴직 대상자'},inplace=True)
parental_leave_ratio.rename(columns={'데이터':'육아 휴직 비율'},inplace=True)
```

# 연령별 육아휴직자의 시간에 따른 변화 확인

```python
df_pivot = pd.pivot_table(parental_leave,index='시점',columns=['연령별'],values=['육아 휴직자'])

df_pivot.plot(kind='line',figsize=(12,8))
# 레이블 및 제목 설정
plt.xlabel('시점')
plt.ylabel('육아휴직자')
plt.title('연령별 육아 휴직자의 시간에 따른 변화')
plt.xticks(rotation=45)
plt.legend(loc='upper right')

plt.show()
```

# 연령별 육아휴직 대상자의 시간에 따른 변화 확인

```python
df_pivot = pd.pivot_table(parental_leave_eligible,index='시점',columns=['연령별'],values=['육아 휴직 대상자'])

df_pivot.plot(kind='line',figsize=(12,8))
# 레이블 및 제목 설정
plt.xlabel('시점')
plt.ylabel('육아휴직 대상자')
plt.title('연령별 육아 휴직 대상자의 시간에 따른 변화')
plt.xticks(rotation=45)
plt.legend(loc='upper right')

plt.show()
```

# 연령별 육아휴직 비율의 시간에 따른 변화 확인

```python
df_pivot = pd.pivot_table(parental_leave_ratio,index='시점',columns=['연령별'],values=['육아 휴직 비율'])

df_pivot.plot(kind='line',figsize=(12,8))
# 레이블 및 제목 설정
plt.xlabel('시점')
plt.ylabel('육아 휴직 비율')
plt.title('연령별 육아 휴직 비율의 시간에 따른 변화')
plt.xticks(rotation=45)
plt.legend(loc='upper right')

plt.show()
```

# 결론
- 30대 이상의 출생아 부모 중 육아 휴직 대상자의 수는 점점 줄어들고 있는 반면, 육아 휴직자는 점점 증가하고 있음
- 30대 미만의 출생아 부모 중 육아 휴직 대상자의 수와 육아 휴직자는 모두 점점 줄어들고 있음
- 육아 휴직 대상자 중 실제 육아 휴직을 하는 비율의 경우 30대 미만이 가장 높았고, 연령이 많아질 수록 육아 휴직을 하는 비율이 낮아짐.
- 하지만, 과거에 비해 육아 휴직을 하는 비율이 점점 증가하고 있다는 점에서 긍정적으로 평가할 수 있음.
